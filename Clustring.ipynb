1. What is unsupervised learning in the context of machine learning?


**Unsupervised learning** in machine learning refers to a type of algorithm that learns patterns from data **without any labeled outcomes or target variables**. Unlike supervised learning, where models are trained on input-output pairs, unsupervised learning explores the underlying structure, relationships, or distribution in the data to find meaningful patterns.

### Key points:

* No labeled data (no explicit correct answers).
* The goal is often to **discover hidden patterns, groupings, or data representations**.
* Common tasks include:

  * **Clustering** (e.g., K-means, hierarchical clustering) ‚Äî grouping similar data points.
  * **Dimensionality reduction** (e.g., PCA, t-SNE) ‚Äî simplifying data while preserving important information.
  * **Anomaly detection** ‚Äî finding unusual data points.

### Example:

* Given customer purchase data without labels, unsupervised learning might group customers into segments based on buying behavior.

2. How does K-Means clustering algorithm work?


Great question! Here‚Äôs a clear explanation of how **K-Means clustering** works:

---

### What is K-Means?

K-Means is an **unsupervised learning algorithm** used to partition data into **K clusters**, where each data point belongs to the cluster with the nearest mean (centroid).

---

### How it works step-by-step:

1. **Initialize**: Choose K initial cluster centroids randomly (or by some heuristic).

2. **Assign**: For each data point, assign it to the cluster whose centroid is closest (usually using Euclidean distance).

3. **Update**: Recalculate the centroids of each cluster as the mean of all data points assigned to that cluster.

4. **Repeat**: Repeat the **Assign** and **Update** steps until:

   * The assignments no longer change (convergence), or
   * A maximum number of iterations is reached.

---

### Summary:

* K-Means tries to minimize the **within-cluster sum of squares** (variance).
* It finds clusters where data points are close to their cluster centroid.
* Sensitive to initial centroids and may converge to local minima.

---

### Example intuition:

If you have points scattered on a plane and want to group them into 3 clusters, K-Means will iteratively move cluster centers and assign points until the best grouping is found.

3. Explain the concept of a dendrogram in hierarchical clustering.


Sure! Here‚Äôs what a **dendrogram** is in the context of **hierarchical clustering**:

---

### What is a dendrogram?

A **dendrogram** is a tree-like diagram that **visualizes the process of hierarchical clustering**, showing how clusters are formed step-by-step by merging (or splitting) data points or clusters.

---

### In hierarchical clustering:

* You start with each data point as its own cluster.
* Iteratively, the two closest clusters are merged.
* This continues until all points are grouped into a single cluster (or until a stopping criterion).

---

### What the dendrogram shows:

* **Leaves (bottom)**: Individual data points or small clusters.
* **Branches**: Mergers of clusters at different levels.
* **Height on the vertical axis**: The distance (or dissimilarity) at which clusters are merged.

---

### How to interpret it:

* The vertical height where two branches join represents the distance between those clusters.
* By "cutting" the dendrogram at a particular height, you decide the number of clusters.
* Clusters that merge low on the dendrogram are more similar than those that merge high.

---

### Visual summary:

```
Height
  |
  |        ________
  |       |        |____
  |    ___|___         |____
  |   |       |___          |
  |___|           |_________|
  +---------------------------> Data points
```

4. What is the main difference between K-Means and Hierarchical Clustering?


Great question! Here's a clear comparison of the **main differences between K-Means and Hierarchical Clustering**:

---

| Feature                    | **K-Means Clustering**                                 | **Hierarchical Clustering**                                  |
| -------------------------- | ------------------------------------------------------ | ------------------------------------------------------------ |
| **Approach**               | Partition-based (flat clustering)                      | Tree-based (builds a hierarchy of clusters)                  |
| **Number of Clusters (K)** | Must be specified before training                      | Not required initially; determined by cutting the dendrogram |
| **Algorithm Type**         | Iterative refinement (centroid-based)                  | Agglomerative (bottom-up) or divisive (top-down)             |
| **Output**                 | Flat clusters (no hierarchy)                           | Dendrogram (shows cluster merging/splitting hierarchy)       |
| **Scalability**            | Faster and more scalable on large datasets             | Slower, especially with large datasets                       |
| **Sensitivity to noise**   | Sensitive to outliers and initial centroids            | More robust to noise/outliers in some variants               |
| **Flexibility**            | Assumes spherical clusters (due to Euclidean distance) | Can capture nested or irregular-shaped clusters              |

---

### Summary:

* **K-Means** is efficient, but needs you to choose `K` and works best with well-separated, spherical clusters.
* **Hierarchical Clustering** builds a tree of clusters and doesn‚Äôt need `K` upfront, but it‚Äôs slower and harder to scale.

5. What are the advantages of DBSCAN over K-Means?


Excellent question! **DBSCAN** (Density-Based Spatial Clustering of Applications with Noise) offers several advantages over **K-Means**, especially for more complex clustering tasks.

---

### ‚úÖ **Advantages of DBSCAN over K-Means:**

| Advantage                            | Explanation                                                                                        |
| ------------------------------------ | -------------------------------------------------------------------------------------------------- |
| **No need to specify K**             | DBSCAN does not require the number of clusters (`K`) to be specified in advance.                   |
| **Identifies arbitrary shapes**      | Can find clusters of **any shape**, not just spherical (which K-Means assumes).                    |
| **Handles noise/outliers**           | Automatically identifies and excludes **noise points** (outliers) as not belonging to any cluster. |
| **Robust to initialization**         | Doesn't depend on random centroids like K-Means does, so it‚Äôs more consistent.                     |
| **Works with varying cluster sizes** | Can detect clusters of **different densities and sizes**, which K-Means struggles with.            |

---

### üîª Limitations to consider:

* DBSCAN may struggle with **very high-dimensional data**.
* Requires tuning of two parameters: `eps` (neighborhood size) and `min_samples` (minimum points in a neighborhood to form a dense region).
* Performance can degrade with **varying densities** in the same dataset.

---

### In short:

* Use **K-Means** when clusters are well-separated, spherical, and `K` is known.
* Use **DBSCAN** when clusters have irregular shapes, contain noise, or `K` is unknown.


6. When would you use Silhouette Score in clustering?


Great question! You would use the **Silhouette Score** to **evaluate the quality of clustering** ‚Äî it helps you understand how well-separated and cohesive your clusters are.

---

### ‚úÖ **What is the Silhouette Score?**

The **Silhouette Score** measures how similar each data point is to its **own cluster** compared to **other clusters**.

It combines:

* **Cohesion**: How close a point is to other points in the same cluster.
* **Separation**: How far a point is from points in the nearest different cluster.

---

### üß† **When to use it:**

| Use Case                                   | Why Use Silhouette Score?                                                           |
| ------------------------------------------ | ----------------------------------------------------------------------------------- |
| ‚úÖ **Choosing the best number of clusters** | Helps determine the optimal value of `K` or number of clusters by comparing scores. |
| ‚úÖ **Comparing clustering algorithms**      | Compare performance of different algorithms (e.g., K-Means vs DBSCAN).              |
| ‚úÖ **Evaluating cluster quality**           | Quantitatively assess if the clusters are well-separated and internally tight.      |
| ‚úÖ **No ground truth labels**               | Ideal when you don‚Äôt have labeled data (unsupervised learning).                     |

---

### üßÆ **Silhouette Score range:**

* **+1**: Point is well matched to its own cluster and far from others.
* **0**: Point lies on the boundary between two clusters.
* **-1**: Point is likely in the wrong cluster.

7. What are the limitations of Hierarchical Clustering?


Great question! While **Hierarchical Clustering** is powerful for visualizing data structure (especially with dendrograms), it has several **limitations** you should be aware of:

---

### ‚ö†Ô∏è **Limitations of Hierarchical Clustering:**

| Limitation                                            | Description                                                                                                                      |
| ----------------------------------------------------- | -------------------------------------------------------------------------------------------------------------------------------- |
| **1. Not scalable for large datasets**                | Time and space complexity are high (`O(n¬≤)` or worse), making it impractical for big data.                                       |
| **2. No undo**                                        | Once clusters are merged (in agglomerative) or split (in divisive), the decisions are final ‚Äî it can‚Äôt go back and fix mistakes. |
| **3. Sensitive to noise and outliers**                | Outliers can drastically affect the hierarchy and lead to misleading clusters.                                                   |
| **4. Requires distance metric**                       | Results depend heavily on the choice of distance metric (e.g., Euclidean, Manhattan).                                            |
| **5. Difficulty choosing optimal number of clusters** | Requires cutting the dendrogram at a certain level, which may be subjective or unclear.                                          |
| **6. No objective function to optimize**              | Unlike K-Means (which minimizes within-cluster variance), hierarchical clustering doesn't optimize a clear, single criterion.    |
| **7. Poor performance with varying density/shape**    | Struggles when clusters have different densities or non-globular shapes.                                                         |

---

### In summary:

Hierarchical clustering is great for **visual insight** and **small to medium-sized datasets**, but less suited for **large, noisy, or complex datasets** where algorithms like **DBSCAN** or **K-Means** may perform better.



8. Why is feature scaling important in clustering algorithms like K-Means?


Excellent question! **Feature scaling is crucial in clustering algorithms like K-Means** because these algorithms rely heavily on **distance calculations** ‚Äî typically **Euclidean distance** ‚Äî to group data points.

---

### ‚úÖ **Why is feature scaling important?**

| Reason                         | Explanation                                                                                                                                                                              |
| ------------------------------ | ---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **1. Distance dominance**      | K-Means computes distances between data points and centroids. If one feature has a larger scale (e.g., income in thousands vs. age in years), it will dominate the distance calculation. |
| **2. Fair contribution**       | Feature scaling ensures all features contribute equally to distance computations ‚Äî preventing bias toward high-magnitude features.                                                       |
| **3. Better cluster quality**  | Unscaled data can lead to **poor cluster formation**, with centroids skewed toward high-scale dimensions.                                                                                |
| **4. More meaningful results** | Scaling helps produce clusters that reflect the actual structure of the data across **all features**, not just the numerically dominant ones.                                            |

---

### üßÆ Common scaling techniques:

* **Standardization** (Z-score): Mean = 0, Std = 1
  ‚Üí `StandardScaler()` in scikit-learn
* **Normalization** (Min-Max): Scales data to \[0, 1]
  ‚Üí `MinMaxScaler()`

---

### üîç Example:

If you cluster customers using `age` (20‚Äì70) and `income` (30,000‚Äì200,000), income will dominate unless scaled ‚Äî K-Means will form clusters primarily based on income, ignoring age.

---

### ‚úÖ Rule of thumb:

> Always scale your features **before using K-Means or any distance-based algorithm** (like DBSCAN, hierarchical clustering, etc.).



9. How does DBSCAN identify noise points?


Great question! **DBSCAN** (Density-Based Spatial Clustering of Applications with Noise) identifies **noise points** as a core part of its clustering logic.

---

### üß† **How DBSCAN identifies noise points:**

A point is labeled as **noise (or outlier)** if it **does not meet the criteria to be part of any cluster**.

DBSCAN uses two key parameters:

* **`eps`**: The radius of the neighborhood around a point.
* **`min_samples`**: The minimum number of points required in the `eps`-neighborhood to form a **dense region** (i.e. a cluster).

---

### üìå Definitions:

* **Core Point**: A point with **at least `min_samples`** within its `eps`-neighborhood.
* **Border Point**: Has **fewer than `min_samples`**, but is in the neighborhood of a core point.
* **Noise Point**: Not a core point and **not reachable** from any core point.

---

### üß≠ Example:

Imagine you have a point `P`:

* If `P` has ‚â• `min_samples` neighbors within radius `eps` ‚Üí it's a **core point**.
* If not, but it‚Äôs close to a core point ‚Üí it‚Äôs a **border point**.
* If it's neither ‚Üí it's marked as **noise**.

---

### üîç Summary:

> DBSCAN marks a point as **noise** if it's **isolated** ‚Äî it‚Äôs not dense enough to form a cluster and isn‚Äôt near a dense region.

10. Define inertia in the context of K-Means.


Great question! In the context of **K-Means clustering**, **inertia** is a key concept that helps evaluate the quality of the clustering.

---

### üìå **Definition:**

**Inertia** is the **sum of squared distances between each data point and the centroid of its assigned cluster**.

Mathematically:

$$
\text{Inertia} = \sum_{i=1}^{k} \sum_{x \in C_i} ||x - \mu_i||^2
$$

Where:

* $k$ is the number of clusters
* $C_i$ is the set of points in cluster $i$
* $\mu_i$ is the centroid of cluster $i$
* $||x - \mu_i||^2$ is the squared Euclidean distance

---

### üí° **Intuition:**

* Inertia measures how **tight** the clusters are.
* **Lower inertia** means data points are **closer to their centroids**, indicating **better-defined clusters**.

---

### üß™ **When is it used?**

* **To evaluate model fit**: You want to minimize inertia.
* **In the Elbow Method**: Plot inertia vs. number of clusters `K` to find the "elbow point" ‚Äî the optimal `K`.

---



11. What is the elbow method in K-Means clustering?


Great question! The **Elbow Method** is a popular technique to determine the **optimal number of clusters (K)** in **K-Means clustering**.

---

### üìå **What is the Elbow Method?**

The Elbow Method involves plotting the **inertia** (within-cluster sum of squared errors) for different values of $K$, and identifying the "elbow point" ‚Äî the point where the rate of decrease sharply slows.

---

### üßÆ **Steps to use the Elbow Method:**

1. Run K-Means clustering for a range of $K$ values (e.g., 1 to 10).
2. Record the **inertia** (sum of squared distances to cluster centroids) for each value of $K$.
3. Plot $K$ (x-axis) vs. inertia (y-axis).
4. Look for the **elbow point**: the value of $K$ after which inertia decreases more slowly.

---

### üìà **Interpretation:**

* Before the elbow: Adding more clusters **significantly improves** the fit (reduces inertia).
* After the elbow: Adding more clusters **yields diminishing returns**.

**The "elbow" = optimal $K$**.

---

### üîç Example Insight:

If your elbow appears at $K = 3$, that suggests **3 clusters** is a good balance between simplicity and fit.

12. Describe the concept of "density" in DBSCAN.


Great question! In **DBSCAN** (Density-Based Spatial Clustering of Applications with Noise), the concept of **"density"** is central to how the algorithm defines clusters.

---

### üìå **What is "density" in DBSCAN?**

In DBSCAN, **density** refers to how many data points fall within a specified **radius** (`eps`) around a point.

A region is considered **dense** if it contains at least a **minimum number of points** (`min_samples`) within that radius.

---

### üß± Two key parameters:

1. **`eps`**: Radius of the neighborhood around a point.
2. **`min_samples`**: Minimum number of points (including the point itself) required to be considered dense.

---

### üìç Types of Points Based on Density:

| Type             | Definition                                                                  |
| ---------------- | --------------------------------------------------------------------------- |
| **Core Point**   | Has **‚â• `min_samples`** points in its `eps`-neighborhood (including itself) |
| **Border Point** | Has < `min_samples` points in `eps`, but is **reachable** from a core point |
| **Noise Point**  | Neither a core point nor a border point ‚Äî not in a dense region             |

---

### üß† Why density matters:

* DBSCAN **groups** core points together if they are close (within `eps`) ‚Äî forming a **cluster**.
* Border points are **attached to clusters**, but not dense enough to form clusters themselves.
* Noise points are **excluded** entirely.

---

### üß≠ Summary:

> In DBSCAN, a **cluster is a connected region of high density**, separated by regions of low density (noise).



13. Can hierarchical clustering be used on categorical data?


Yes, **hierarchical clustering can be used on categorical data**, but it requires **special handling**, because traditional distance metrics like **Euclidean distance** do **not work well** with categorical variables.

---

### üß© Why Euclidean distance fails:

Euclidean distance assumes **numerical** values and meaningful arithmetic operations. Categorical variables (like ‚Äúred‚Äù, ‚Äúblue‚Äù, ‚Äúgreen‚Äù) don‚Äôt have a numeric scale, so their distances can‚Äôt be directly computed.

---

### ‚úÖ **How to apply hierarchical clustering on categorical data:**

1. **Use a suitable distance metric** for categorical data:

   * **Hamming distance**: Measures the number of positions at which the corresponding elements are different.
   * **Jaccard distance**: Useful for binary or set-like data.
   * **Gower distance**: Can handle mixed data types (numerical + categorical).

2. **Use appropriate linkage methods**:

   * You can use standard linkage methods (e.g., average, complete, single), as long as your **distance matrix** is properly computed.

3. **Use specialized libraries/tools**:

   * `scipy.cluster.hierarchy` works with custom distance matrices.
   * Libraries like `kmodes` or `gower` in Python are specifically designed for clustering categorical data.

---

### üîç Example: Hamming Distance



````python
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

data = [['Male', 'Yes'], ['Female', 'No'], ['Female', 'Yes'], ['Male', 'No']]
encoded_data = [[0 if val == 'Male' or val == 'Yes' else 1 for val in row] for row in data]

# Compute Hamming distance
dist_matrix = pdist(encoded_data, metric='hamming')

# Perform hierarchical clustering
linkage_matrix = linkage(dist_matrix, method='average')

# Plot dendrogram
dendrogram(linkage_matrix)
plt.show()
````

14. What does a negative Silhouette Score indicate?


Great question! A **negative Silhouette Score** indicates that a data point is likely **misclassified** or **assigned to the wrong cluster**.

---

### üß† What does it mean?

* The Silhouette Score for a point is calculated as:

$$
s = \frac{b - a}{\max(a, b)}
$$

Where:

* $a$ = average distance between the point and other points in the **same cluster** (cohesion)

* $b$ = average distance between the point and points in the **nearest neighboring cluster** (separation)

* If **$a > b$**, the score $s$ becomes **negative**.

---

### üö© Interpretation of negative score:

* The point is **closer on average to points in another cluster** than to points in its own cluster.
* This means the clustering algorithm likely **assigned the point to the wrong cluster**.
* A cluster with many points having negative scores suggests **poor clustering quality**.

---

### üßÆ Range of Silhouette Scores:

| Score Range    | Meaning                                |
| -------------- | -------------------------------------- |
| Close to +1    | Well-clustered, appropriate assignment |
| Around 0       | On the boundary between clusters       |
| Negative (< 0) | Likely misclassified or noisy point    |

15. Explain the term 'linkage criteria" in hierarchical clustering.


Absolutely! The term **"linkage criteria"** is fundamental in **hierarchical clustering** because it defines **how to measure the distance between clusters** when deciding which clusters to merge.

---

### üìå What is **Linkage Criteria**?

* When hierarchical clustering merges two clusters, it needs a rule to compute the **distance between clusters** (not just between individual points).
* The **linkage criteria** determines **how this cluster-to-cluster distance is calculated**.

---

### üß± Common linkage methods:

| Linkage Type         | Description                                                                              |
| -------------------- | ---------------------------------------------------------------------------------------- |
| **Single Linkage**   | Distance between the **closest pair of points** in the two clusters (minimum distance).  |
| **Complete Linkage** | Distance between the **farthest pair of points** in the two clusters (maximum distance). |
| **Average Linkage**  | Average distance between **all pairs of points** in the two clusters.                    |
| **Ward‚Äôs Linkage**   | Minimizes the **increase in total within-cluster variance** after merging clusters.      |

---

### üß† How it affects clustering:

* **Single linkage** can cause **"chaining"**, where clusters become long and snake-like.
* **Complete linkage** tends to create **compact, spherical clusters**.
* **Average linkage** is a compromise between single and complete.
* **Ward‚Äôs method** is often preferred for its focus on minimizing variance and tends to produce balanced clusters.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

Great question! **K-Means clustering can struggle with datasets that have clusters of varying sizes or densities** because of how the algorithm fundamentally works.

---

### üß† Why K-Means struggles with varying cluster sizes or densities:

1. **Assumes spherical, equal-sized clusters**
   K-Means partitions data into clusters by minimizing the sum of squared distances to cluster centroids, effectively assuming clusters are roughly **spherical and similar in size**. When cluster sizes vary widely, centroids can be pulled toward larger clusters, misclassifying points in smaller clusters.

2. **Uses Euclidean distance which favors compact clusters**
   K-Means relies on Euclidean distance which favors **compact, evenly dense** clusters. Sparse or elongated clusters can be split incorrectly or merged with denser clusters.

3. **Sensitive to cluster density differences**
   Clusters with different densities can cause K-Means to assign points from sparse clusters to nearby dense clusters, as centroids are influenced more by denser regions.

4. **Fixed number of clusters $K$**
   When clusters vary in size and density, picking a fixed $K$ may not reflect the true structure, resulting in poor clustering.

---

### üß≠ Example:

Imagine two clusters:

* One is a small, tight group.
* The other is a large, spread-out group.

K-Means may assign some points from the small cluster incorrectly because the centroid of the large cluster dominates the space.

---

### üîë Summary:

> K-Means performs best when clusters are **well-separated, similar in size, and roughly spherical**. It struggles with **varying sizes/densities**, where density-based methods like **DBSCAN** work better.



17. What are the core parameters in DBSCAN, and how do they influence clustering?


Great question! The two core parameters in **DBSCAN** that control how clustering works are **`eps`** and **`min_samples`**.

### üîë Core Parameters in DBSCAN:

18. How does K-Means++ improve upon standard K-Means initialization?




### üîç How standard K-Means initialization works:

* It **randomly selects** $K$ initial centroids from the dataset.
* Poor random initialization can lead to **bad clusters** or slow convergence.
* Results can vary widely depending on the initial centroids.

---

### üöÄ What K-Means++ does differently:

1. **Selects the first centroid randomly** from the data points.
2. For each subsequent centroid, it selects a point **with probability proportional to its squared distance** from the nearest existing centroid.

   * This means points **farther away from existing centroids** are more likely to be chosen.
3. This spreads out the initial centroids **more evenly** across the data.

---

### üí° Benefits of K-Means++:

* **Better starting points**: Initial centroids are well spread out.
* **Faster convergence**: Requires fewer iterations to reach a good solution.
* **Improved clustering quality**: More likely to find a global optimum or a better local optimum.

---

### Summary:

| Method    | Initialization                 | Outcome                     |
| --------- | ------------------------------ | --------------------------- |
| K-Means   | Random selection               | Potentially poor clusters   |
| K-Means++ | Distance-based smart selection | Better, more stable results |

19. What is agglomerative clustering?


Sure! Here‚Äôs a clear explanation:

---

### üìå What is **Agglomerative Clustering**?

**Agglomerative clustering** is a type of **hierarchical clustering** that builds a hierarchy of clusters by **starting with each data point as its own cluster** and **iteratively merging the closest clusters** until all points belong to a single cluster or a stopping criterion is met.

---

### How it works step-by-step:

1. **Start:** Each data point is considered a **single cluster**.
2. **Compute distances:** Calculate the distance between every pair of clusters.
3. **Merge:** Find the two closest clusters and merge them into one.
4. **Repeat:** Update the distance matrix and repeat merging until:

   * Only one cluster remains, or
   * A desired number of clusters is achieved.

---

### Key points:

* It‚Äôs a **bottom-up** approach (starting with single points and building up).
* The merging decision depends on the **linkage criteria** (e.g., single, complete, average linkage).
* Results can be visualized with a **dendrogram**, showing cluster merges at different levels.

---

### Contrast with divisive clustering:

* Agglomerative: bottom-up merging.
* Divisive: top-down splitting.


20. What makes Silhouette Score a better metric than just inertia for model evaluation?

Great question! Both **Silhouette Score** and **inertia** are popular metrics for evaluating clustering, but they capture different aspects and Silhouette Score often provides a more nuanced evaluation.

---

### Why Silhouette Score can be better than inertia:

| Aspect                              | Inertia                                                                                   | Silhouette Score                                                                    |
| ----------------------------------- | ----------------------------------------------------------------------------------------- | ----------------------------------------------------------------------------------- |

| **Robustness to cluster shape**     | Assumes spherical clusters due to Euclidean distance                                      | Can handle various cluster shapes better due to pairwise distance evaluation        |
| **Used for determining optimal K?** | Often used with elbow method, which is subjective                                         | Provides a direct score for different K, easier to choose optimal                   |

---

### Intuition:

* **Inertia** only tells you how close points are to their own cluster centroid.
* **Silhouette Score** tells you how well points fit their own cluster compared to other clusters ‚Äî considering **both cohesion and separation**.

---

### Summary:

> Silhouette Score gives a **more balanced and interpretable evaluation** of clustering quality, especially useful when clusters have irregular shapes or are not well separated.

````python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with 4 centers
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plot the data points colored by cluster assignment
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=30, cmap='viridis', alpha=0.6)

# Plot cluster centroids
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')

plt.title('K-Means Clustering on Synthetic Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
````

29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.


````python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data
X, y = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=42)

# Evaluate silhouette scores for k = 2 to 5
k_values = range(2, 6)
scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

# Plot the silhouette scores as a bar chart
plt.figure(figsize=(8, 5))
plt.bar(k_values, scores, color='skyblue')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for k = 2 to 5')
plt.xticks(k_values)
plt.show()
````


31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.


````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from matplotlib.colors import ListedColormap

# Generate overlapping data
X, y = make_blobs(n_samples=500,
                  centers=3,
                  cluster_std=1.8,  # Higher std for overlap
                  random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Create mesh grid for decision boundary
h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# Predict cluster for each point in mesh
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundaries and clusters
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='Pastel1')
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='Set1', s=40, edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='black', marker='X', label='Centroids')

plt.title('K-Means Clustering with Decision Boundaries (Overlapping Clusters)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
````

44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.


````python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

# Load Iris dataset
iris = load_iris()
X = iris.data

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X)

# Compute Silhouette scores for each sample
silhouette_vals = silhouette_samples(X, cluster_labels)

# Sort scores within each cluster for better visualization
y_lower = 10
plt.figure(figsize=(8, 6))

for i in range(3):
    ith_cluster_silhouette_vals = silhouette_vals[cluster_labels == i]
    ith_cluster_silhouette_vals.sort()

    size_cluster = ith_cluster_silhouette_vals.shape[0]
    y_upper = y_lower + size_cluster

    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_vals,
                      alpha=0.7)

    plt.text(-0.05, y_lower + 0.5 * size_cluster, str(i))
    y_lower = y_upper + 10  # 10 for spacing between clusters

plt.xlabel('Silhouette Coefficient')
plt.ylabel('Sample Index')
plt.title('Silhouette Coefficient for each sample (Iris, KMeans)')
plt.axvline(x=silhouette_score(X, cluster_labels), color='red', linestyle='--', label='Average Silhouette Score')
plt.legend()
plt.show()
````

46. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features).


````python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import pandas as pd

# Load Wine dataset
wine = load_wine()
X = wine.data[:, :4]  # first 4 features
feature_names = wine.feature_names[:4]

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)

# Prepare DataFrame for seaborn
df = pd.DataFrame(X, columns=feature_names)
df['Cluster'] = labels.astype(str)  # convert to string for coloring

# Plot pairplot with clusters
sns.pairplot(df, hue='Cluster', palette='Set1', diag_kind='kde')
plt.suptitle('KMeans Cluster Assignments on Wine Dataset (First 4 Features)', y=1.02)
plt.show()
````

47. Generate noisy blobs using make blobs and use DBSCAN to identify both clusters and noise points. Print the count

````python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Generate noisy blobs
X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=1.5, random_state=42)

# Scale the data (important for DBSCAN)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Count clusters (excluding noise, label == -1)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print(f"Number of clusters found: {n_clusters}")
print(f"Number of noise points detected: {n_noise}")
````

48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters

````python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering

# Load Digits dataset
digits = load_digits()
X = digits.data

# Reduce dimensions to 2D using t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Apply Agglomerative Clustering
agg_clust = AgglomerativeClustering(n_clusters=10, linkage='ward')
labels = agg_clust.fit_predict(X_tsne)

# Plot clusters
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='tab10', s=50, edgecolor='k')
plt.title('Agglomerative Clustering on Digits Dataset (t-SNE Reduced)')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.colorbar(scatter, ticks=range(10), label='Cluster Label')
plt.show()
````
