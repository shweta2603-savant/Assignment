Q 5.Generate a synthetic dataset with correlated features.Visulize the correlation matrix of a dataset using a heatmap.

To generate a synthetic dataset with correlated features and visualize the correlation matrix using a heatmap, we'll follow these steps:

Generate the synthetic dataset: We will use Python and the NumPy library to generate random data with correlated features. The features will be constructed such that their correlation is controlled.

Visualize the correlation matrix: We'll calculate the correlation matrix using pandas and plot it using seaborn's heatmap functionality.

Here's the full Python code to achieve this:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Number of samples and features
n_samples = 500
n_features = 5

# Generate random data
# Let's create a correlation matrix for the features
corr_matrix = np.array([
    [1.0, 0.8, 0.5, 0.3, 0.1],  # Feature 1 correlations
    [0.8, 1.0, 0.6, 0.4, 0.2],  # Feature 2 correlations
    [0.5, 0.6, 1.0, 0.7, 0.4],  # Feature 3 correlations
    [0.3, 0.4, 0.7, 1.0, 0.6],  # Feature 4 correlations
    [0.1, 0.2, 0.4, 0.6, 1.0],  # Feature 5 correlations
])

# Use a multivariate normal distribution to generate correlated features
mean = np.zeros(n_features)  # mean of 0 for each feature
cov = corr_matrix  # covariance matrix defined above

# Generate the synthetic dataset
data = np.random.multivariate_normal(mean, cov, size=n_samples)

# Convert the data to a pandas DataFrame
df = pd.DataFrame(data, columns=[f'Feature {i+1}' for i in range(n_features)])

# Calculate the correlation matrix of the dataset
corr_df = df.corr()

# Plot the heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_df, annot=True, cmap='coolwarm', center=0, linewidths=1, fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Explanation of the Code:
Generate the Data:

We create a custom correlation matrix (corr_matrix) that defines the correlation between the features. The diagonal is set to 1, and the off-diagonal elements specify the correlation between pairs of features.
We then use the np.random.multivariate_normal function to generate random data that follows this correlation structure.
Create the DataFrame:

The generated data is stored in a pandas DataFrame for easier handling and visualization.
Calculate the Correlation Matrix:

The df.corr() method computes the Pearson correlation coefficient matrix for the DataFrame.
Plot the Heatmap:

The seaborn.heatmap function visualizes the correlation matrix with annotations and a color map. The annot=True argument ensures that the correlation values are displayed on the heatmap.
Expected Output:
This code will generate a heatmap that visually represents the correlations between features. The features that are more highly correlated will have darker colors on the heatmap (if the correlation is close to 1 or -1), while weaker correlations will appear in lighter shades.

If you run this code on your local environment (make sure to have the necessary libraries installed), you'll see a heatmap with correlations between the features, showing the relationships within the synthetic dataset.

Q 5.Generate a synthetic dataset with correlated features.Visualize the correlation matrix of a dataset using a heatmap.

Steps to Follow:
Generate the synthetic dataset with correlated features using Python.
Visualize the correlation matrix with a heatmap.
Here’s the code you can use:



```python
# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate synthetic dataset with correlated features
# Define the number of data points
n_samples = 100

# Generate random data with some correlation
np.random.seed(42)
X1 = np.random.normal(0, 1, n_samples)  # Feature 1: Random normal distribution
X2 = 0.8 * X1 + np.random.normal(0, 0.5, n_samples)  # Feature 2: Correlated with Feature 1
X3 = 0.5 * X1 + 0.5 * X2 + np.random.normal(0, 0.5, n_samples)  # Feature 3: Correlated with Feature 1 and Feature 2
X4 = np.random.normal(0, 1, n_samples)  # Feature 4: Independent from others

# Create a DataFrame
data = pd.DataFrame({
    'Feature1': X1,
    'Feature2': X2,
    'Feature3': X3,
    'Feature4': X4
})

# Step 2: Compute the correlation matrix
corr_matrix = data.corr()

# Step 3: Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

Explanation:
Generate Synthetic Data:

X1: A standard normal distribution.
X2: Correlated with X1 by a factor of 0.8, with some added noise.
X3: A combination of X1 and X2, introducing a second level of correlation.
X4: Independent feature with no correlation to the others.
Correlation Matrix:

The corr() method computes the Pearson correlation coefficients between each pair of features in the DataFrame.
Heatmap:

sns.heatmap() is used to visualize the correlation matrix as a heatmap, with annotations showing the correlation coefficients.
Output:
The heatmap will show how correlated each pair of features is. The values will range between -1 (perfect negative correlation) to +1 (perfect positive correlation). Features that are highly correlated will have values close to 1 or -1, while features with no correlation will have values close to 0.

Q 2.Create a Bokeh scatter plot using randomly generated x and y values.Use different sizes and colors for the markets based on the 'sizes' and 'colors' columns.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral11

# Generate random data
np.random.seed(42)  # For reproducibility
n = 100  # Number of points
x = np.random.random(n) * 10  # x values
y = np.random.random(n) * 10  # y values
sizes = np.random.randint(5, 50, size=n)  # Random sizes for the points
colors = np.random.choice(Spectral11, n)  # Random colors from a predefined palette

# Create a DataFrame to organize the data
data = pd.DataFrame({
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
})

# Convert DataFrame to ColumnDataSource (Bokeh's data structure)
source = ColumnDataSource(data)

# Create a scatter plot
output_file("scatter_plot.html")  # Specify the output HTML file
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y')

# Add scatter renderer with sizes and colors
p.scatter(x='x', y='y', size='sizes', color='colors', alpha=0.6, source=source)

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.palettes import Category10

# Data
fruits = ['Apples', 'Orange', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Output file
output_file("fruit_counts.html")

# Create a figure
p = figure(x_range=fruits, title="Fruit Count", toolbar_location=None, tools="")

# Add a vertical bar renderer
p.vbar(x=fruits, top=counts, width=0.5, color=Category10[4], legend_field="Fruit")

# Set the labels and title
p.xaxis.axis_label = "Fruits"
p.yaxis.axis_label = "Count"
p.legend.title = "Fruits"

# Customize the appearance
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.location = "top_right"

# Show the plot
show(p)
```

Q 4.Create a Bokeh histogram to visualize the distrubution of the given data.

data_hist = np.random.randn(1000)

hist,edges = np.histogram(data_hist,bins=30)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColumnDataSource

# Generate data
data_hist = np.random.randn(1000)

# Compute histogram
hist, edges = np.histogram(data_hist, bins=30)

# Prepare data for Bokeh
source = ColumnDataSource(data=dict(top=hist, left=edges[:-1], right=edges[1:], bottom=[0]*len(hist)))

# Create a figure
p = figure(title="Histogram",
           x_axis_label='Value',
           y_axis_label='Frequency',
           tools="pan,box_zoom,reset")

# Add a quad glyph for the histogram
p.quad(top='top', bottom='bottom', left='left', right='right', source=source,
       fill_color="skyblue", line_color="white", alpha=0.7)

# Show the plot
output_file("histogram.html")
show(p)
```

Step 1: Import necessary libraries

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
from bokeh.layouts import column

# Generate the data
data_heatmap = np.random.rand(10, 10)

x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

xx, yy = np.meshgrid(x, y)

# Output to notebook
output_notebook()
```

Step 2: Set up the heatmap plot

```python
# Create a color mapper using Viridis color palette
color_mapper = linear_cmap(field_name='value', palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

# Set up the figure
p = figure(width=400, height=400, title="Heatmap", toolbar_location=None,
           x_range=(0, 1), y_range=(0, 1))


# Hide grid lines and axes
p.xgrid.visible = False
p.ygrid.visible = False
p.xaxis.visible = False
p.yaxis.visible = False

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
